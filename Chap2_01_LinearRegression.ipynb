## 01 線形回帰

ポイント

* sklearn  の  LinearRegression クラスを用いて動かすことができる



```
from sklearn.linear_model import LinearRegression

X = [[10.0], [8.0], [13.0], [9.0], [11.0], [14.0],
     [6.0], [4.0], [12.0], [7.0], [5.0]]
y = [8.04, 6.95, 7.58, 8.81, 8.33, 9.96,
     7.24, 4.26, 10.84, 4.82, 5.68]
model = LinearRegression()
model.fit(X, y)
print(model.intercept_)
print(model.coef_)
sampleX = [[0], [1]]
y_pred = model.predict(sampleX)
print(y_pred)
```

サンプルプログラムは上記まで書かれており、出力している変数の意味は下記のように解説がある

* model.intercept_ には切片
* model.coef_ には傾きが入っている

ここから下は、上記のサンプルプログラムにはないが、上記のデータを可視化する。

```
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(
    111,
    xlim=(-2, 15), ylim=(0, 13))
ax.plot(X, y, "bo", label="training data")
ax.plot(sampleX, y_pred, "r+", label="predicted data")

converged_X = [i / 10 for i in range(-20, 150)]
converged_Y = [model.coef_ * i /10 + model.intercept_ for i in range(-20,150)] 
ax.plot(converged_X, converged_Y, "--", label="converged linear regression")
ax.legend()
```
